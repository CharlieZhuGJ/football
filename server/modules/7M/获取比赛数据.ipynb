```python
def get_data_url(gid):
    base = "https://analyse.7m.com.cn/{}/data/".format(gid)
    url_dict = {
        "info" : base + "gameinfo_gb.js",
        "history" : base + "gamehistory_gb.js",
        "teamhistory" : base + "gameteamhistory_gb.js",
        "stats" : base + "gameteamstats_gb.js",
        "fixture" : base + "gameteamfixture_gb.js",
        "prediction" : base + "gameprediction_gb.js",
        "lineup" : base + "gamelineup_gb.js",
        "standings" : base + "gamestandings_gb.js",
        "oddsway" : base + "gameoddsway_gb.js"
    }
    return url_dict
```
