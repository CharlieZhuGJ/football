```python
import json
import urllib
import urllib.request
```

```python
news = "http://news.7m.com.cn/data/game/4074677_json.js"
reviews = "http://report.7m.com.cn/data/game/gb/4074677.js"
injure = "http://data.7m.com.cn/analyse/gb/nline_up_injure/4074677.js"
crowns = "http://crowns2.7m.com.cn/report/4074677.js"
lottery = "http://data.lottery.7m.com.cn/data/4074677.js"
```

```python
def get_data_by_url(url):
    req = urllib.request.urlopen(url)
    # print(req.read().decode())
    return req.read().decode()
```

```python
analyse_news = get_data_by_url(news)
analyse_news = json.loads(analyse_news.split("=")[1][:-1])
analyse_news
```

```python
reviews = get_data_by_url(reviews)
```

```python
reviews = ";\n" + reviews
```

```python
reviews.split(";\nvar ")
```

```python
res = {}
for item in reviews.split(";\nvar "):
    if item:    
        print(item)
        print("++++++++++++++++++++++++")
        kv = item.split("=")
        key = kv[0].strip()
        val = kv[1].strip()
        if val.endswith(";"):
            val = val[:-1]
        res[key] = json.loads(val)
```

```python
res
```

```python
def get_injure(url):
    """
    获取伤停相关信息
    :param url:
    :return:
    """
    res = {}
    injures = get_data_by_url(url)
    injures.replace("\ufeff", "\r\n").split("\r\nvar ")

    for injure in injures:
        if injure:
            kv = injure.split("=")
            key = kv[0].strip()
            val = kv[1].strip().split(";")[0]
            res[key] = val

    return res
```
